```python
import pandas as pd
```

```python
producao_df= pd.read_excel('producao.xlsx')
display(producao_df)
```

```python
#o maior produtor opep/não opep
display(producao_df.max())
```

```python
#dividindo os blocos opep e não opep
opep_df = producao_df.loc[producao_df['BLOCO'] == 'OPEP']
display(opep_df)
```

```python
nopep_df = producao_df.loc[producao_df['BLOCO'] == 'NÃO OPEP']
display(nopep_df)
```
